**Removing Outliers**

```python
numeric_df=df.select_dtypes(include=[np.number])
non_numeric_df=df.select_dtypes(exclude=[np.number])
Q1=numeric_df.quantile(0.25)
Q3=numeric_df.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
numeric_df.shape
```

```python
numeric_df=numeric_df[~((numeric_df<(Q1-1.5*IQR)) | (numeric_df>(Q1+1.5*IQR))).any(axis=1)]
numeric_df.shape
```

```python
df=pd.concat([numeric_df,non_numeric_df],axis=1)
```

**Top 40 Makes by Number of Cars**

```python
df.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Number of Make by Cars")
plt.xlabel('Make')
plt.ylabel('Number of Cars')
plt.show()
```

**Heatmap with Correlations**

```python
plt.figure(figsize=(10,5))
c=numeric_df.corr()
sns.heatmap(c,cmap='BrBG',annot=True)
c
```

**Scatter Plot**

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.scatterplot(x=df['Hp'],y=df['Price'],ax=ax)
ax.set_title("Hp Vs Price")
ax.set_xlabel('Hp')
ax.set_ylabel('Price')
```
